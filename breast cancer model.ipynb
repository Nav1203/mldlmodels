```python
x=b_df.drop(['id','unif_cell_shape','class'],axis=1)
y=b_df['class']
```

```python
x
```

```python
for i in x.columns:
    print(i,x[i].unique())
```

```python
for i in x['bare_nuclei'].unique():
    if type(i)!=int and i!='?':
        x["bare_nuclei"].replace({i: int(i)}, inplace=True)
    else:
        x["bare_nuclei"].replace({i: 0}, inplace=True)
    
```

```python
x['bare_nuclei'].unique()
```

```python
a=x['bare_nuclei'].mean()
a
```

```python
x['bare_nuclei'].replace({0:a},inplace=True)
```

```python
x
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
maximum=0
for i in range(0,12):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    bst_model=LogisticRegression(C=1e6,max_iter=1e9,random_state=0)
    bst_model.fit(xtrain,ytrain)
    ##print(bst_model.score(xtest,ytest),i)
    if bst_model.score(xtest,ytest)>maximum:
        maximum=bst_model.score(xtest,ytest)
        max_iter=i
print(max_iter,maximum)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=3)
cancer_model=LogisticRegression(C=1e6,max_iter=1e9,random_state=0)
cancer_model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
predicted=cancer_model.predict(xtest)
confusion_matrix(ytest,predicted)
```

```python
print(classification_report(ytest,predicted))
```
