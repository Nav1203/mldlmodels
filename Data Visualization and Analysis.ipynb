```python
import matplotlib.pyplot as plt
# Descriptive Plots - Boxplot, Density etc.
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
bst_data=pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/boston_train.csv")
bst_data
```

```python
bst_data.notnull().sum()
```

**LINE PLOT**
Use line plot for univariate data analysis (continous)

```python
plt.figure(figsize=(18,8))
plt.plot(bst_data['AGE'])
plt.title("Age vs Rows Line Plot", fontsize = 10)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

**BAR PLOT**
    Bivariate visualization - X must be discrete and Y can be discrete or continious

```python
plt.figure(figsize=(18,8))
plt.bar(x=bst_data['ZN'],height = bst_data['AGE'],width=1.5,color='deeppink')
plt.title("Age vs Rows Line Plot", fontsize = 10)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show() 
```

**Histogram**
    Frequency of data in an interval-variance data
    Univariate Analysis of Continous

```python
plt.figure(figsize=(18,8))
plt.hist(bst_data['MEDV'],bins=15,color = 'darkviolet')
plt.title("Histogram", fontsize = 10)
plt.xlabel("MEDV")
plt.ylabel("Frequency")
plt.show() 
```

**Scatter Plot**
    Bivariate Analysis - Relationship Plot between two features

```python
ir_data=pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/iris.csv")
```

```python
ir_data
```

```python
ir_data['species'].unique()
```

```python
plt.figure(figsize=(13,8))
plt.scatter(x=ir_data['sepal_length'],y=ir_data['petal_length'],c='red')
plt.title('Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(13,8))
plt.scatter(x=ir_data['sepal_length'],y=ir_data['petal_length'],c='red', marker='*',label='petal_length')
plt.scatter(x=ir_data['sepal_width'],y=ir_data['petal_width'],c='m',marker='o',label='petal_width')
plt.title('Scatter Plot')
plt.xlabel('Sepal Length/Sepal Width')
plt.ylabel('Petal Length/Petal Width')
plt.legend()
plt.show()
```

**Seaborn - Scatterplot**

```python
sns.set_style("whitegrid")
plt.figure(figsize=(13,8))
sns.scatterplot(x='sepal_length', y='petal_length',data=ir_data, hue ='species',palette='plasma')
plt.show()
```

**Pie chart**

```python
ir_data['species'].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.pie(ir_data['species'].value_counts(), autopct='%0.2f%%',labels = ir_data['species'].unique(),explode=(0,0,0.1))
plt.show()
plt.savefig('piechart.jpg')
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,8))
sns.distplot(bst_data.MEDV,bins=15)
```

**BOXPLOT**

```python
plt.style.use('ggplot')
plt.figure(figsize=(11,8))
sns.boxplot(data=ir_data, orient='h',palette='plasma')
```
