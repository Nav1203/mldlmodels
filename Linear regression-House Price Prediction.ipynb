```python
import matplotlib.pyplot as plt
# Descriptive Plots - Boxplot, Density etc.
import seaborn as sns
import pandas as pd
import numpy as np
#using inbuilt datasets
from sklearn.datasets import load_boston
```

```python
print(boston.DESCR)
```

```python
b_df=pd.DataFrame(boston.data, columns=boston.feature_names)
b_df
```

```python
b_df['MEDV']=boston.target
```

**Features**

```python
#tax and rad has multi colinearity so eliminating the weakest i.e rad
##Feature selection
x=b_df.drop(['MEDV','RAD','CHAS'],axis=1)
y=b_df['MEDV']
x
```

```python
x
```

**DATA PREPROCESSING**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
x_scaled=scaler.fit_transform(x)
x_scaled
```

**CROSS VALIDATION**
Splitting Data into training and testing randomly

```python
from sklearn.model_selection import train_test_split
# choose random state to randomly select data or randomly sample data
X_train,X_test,Y_train,Y_test=train_test_split(x_scaled, y, test_size=0.2, random_state=3)
```

**FITTING**

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression(normalize=True)
```

```python
#fit()-training of linear regression model
lin_reg.fit(X_train,Y_train)
```

**ACCURACY**

```python
#R-squared of Test model
lin_reg.score(X_test, Y_test)
```

**Metrics of LR**

```python
#predicted o/p
yhat=lin_reg.predict(X_test)
```

```python
#r2_score,mean_squared_error,mean_absolute_error
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
#coefficent of determination : variance of yhat and ytest
#r2_score is having range (0,1)
r2_score(Y_test,yhat)
```

```python
#MAE (mean absolute error)
mean_absolute_error(Y_test, yhat)
```

```python
mse=mean_squared_error(Y_test,yhat)
mse
```

```python
#RMSE
np.sqrt(mse)
```
