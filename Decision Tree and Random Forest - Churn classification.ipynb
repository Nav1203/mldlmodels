# Data Augmentation

```python
# resample uses "bootstrapping" to regenerate samples by randomly selecting data for each class
from sklearn.utils import resample
df_0=churn_df[churn_df['Churn']== False]
df_1=churn_df[churn_df['Churn']==True]
```

```python
df_1_upsample=resample(df_1,n_samples=288,replace=True,random_state=123)
churn_df1=pd.concat([df_0,df_1_upsample])
churn_df1
```

```python
churn_df1['Churn'].value_counts()
```

```python
x=churn_df1.drop(['Churn','State','Number vmail messages','Total day charge','Total eve charge','Total night charge','Total intl charge','Churn'],axis=1)
y=churn_df1['Churn']
```

```python
display(x)
display(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_tree=DecisionTreeClassifier(criterion="entropy",random_state=0,max_depth=9)
```

```python
clf_tree.fit(xtrain,ytrain)
```

```python
clf_tree.score(xtest,ytest)
```

```python
clf_tree.score(xtrain,ytrain)
```

If training accuracy more than testing than overfitting has taken place
To avoid overfitting
- Post pruning method
- Use ensemble learning-Random Forest

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_rf=RandomForestClassifier(bootstrap=True,max_depth=9,n_estimators=50,random_state=0,criterion='entropy')
```

```python
clf_rf.fit(xtrain,ytrain)
```

```python
clf_rf.score(xtrain,ytrain)
```

```python
clf_rf.score(xtest,ytest)
```

```python
predict=clf_rf.predict(xtest)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(ytest,predict)
```

```python
sns.heatmap(confusion_matrix(ytest,predict),annot=True,fmt='0.0f')
```

```python
print(classification_report(ytest,predict))
```
