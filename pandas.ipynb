**Filtering Data using And and OR operation**

```python
movies_df['Year']==2008
```

```python
movies_df[movies_df['Year']==2008]
```

```python
movies_df[movies_df['Audience score %']==89]
```

```python
#or operation
movies_df[(movies_df['Audience score %']==89) | (movies_df['Genre']=='Comedy')]
```

```python
#and operation
movies_df[(movies_df['Audience score %']>80) & (movies_df['Genre']=='Comedy')]
```

**Slicing using loc and iloc**
Both are used for slicing the Data

```python
# Before ',', it slices rows and after ',' it slices columns using labels
gg=movies_df.loc[10:20,:]
gg
```

```python
gg=movies_df.loc[10:20,'Genre':'Rotten Tomatoes %']
gg
```

```python
movies_df.loc[10,'Genre']
```

```python
# iloc - it uses index position to slice data
movies_df.iloc[10:20,2:7]
```

```python
movies_df.iloc[10:20,-1]
```
